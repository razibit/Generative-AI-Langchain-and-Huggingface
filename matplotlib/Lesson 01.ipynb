```python
import matplotlib.pyplot as plt
```

```python
x = [3,6,9,11,13]
y = [11,2,5,9,15]
x1 = [11,2,5,9,15]
y1 = [3,6,9,11,13]

plt.title("Sales over Time")
plt.plot(x,y, color='cyan',linestyle='-.',marker='*',label="new")
plt.plot(x1,y1, color='black',linestyle='-.',marker='o',label="old")


plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Test Plot")
plt.legend()


plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.plot(x,y1, color='red', label = "Squared")
plt.plot(x,y2, color='blue', label = 'linear')

plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Multiple Lines Plot")
plt.legend()

plt.show()
```
