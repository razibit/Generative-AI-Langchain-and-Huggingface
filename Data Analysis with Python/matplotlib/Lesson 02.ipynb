````python
import matplotlib.pyplot as plt
````

````python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x,y, color='purple', marker = 'o')

plt.title("Customised Line Plot", fontsize=16)
plt.xlabel("X Axis Label", fontsize=14)
plt.ylabel("Y Axis Label", fontsize = 14)

plt.show()
````

````python
plt.plot(x,y, color='purple',marker='o')
plt.title("PLot with Grid")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.grid(True)
plt.show()
````

````python
plt.plot(x,y, color='purple', marker='o')

plt.title("Annotated Plot")

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.annotate("Highest Point", xy=(5,25),xytext=(3,20),arrowprops=dict(facecolor='green',arrowstyle="->"))
plt.show()
````

````python
plt.plot(x,y, color='green',marker='o')
plt.annotate("Lowest Point",xy=(1,1),xytext=(6,1.5),fontsize=12,color='red',arrowprops=dict(facecolor='green',arrowstyle="]-"))

plt.annotate("Middle Point",xy=(3,9),xytext=(3,15),fontsize=12,color='black',arrowprops=dict(facecolor='green',arrowstyle="->"))

plt.grid(True)
plt.show()
````


**Some common arrowstyle options we can use for annotations in Matplotlib:** 

```
'-': Simple line
'->': Arrow
'-[': Bracket
']-': Bracket
'|-|': Bar
'<-': Left arrow
'<->': Left and right arrow
'<|-|>': Left and right bar
'fancy': Fancy arrow
'simple': Simple arrow
'wedge': Wedge arrow
```

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, color='purple', marker='o')

plt.annotate("Arrow", xy=(5, 25), xytext=(3, 20),
             arrowprops=dict(facecolor='black', arrowstyle="->"))
plt.annotate("Fancy", xy=(4, 16), xytext=(2, 10),
             arrowprops=dict(facecolor='red', arrowstyle="fancy"))
plt.annotate("Simple", xy=(3, 9), xytext=(1, 5),
             arrowprops=dict(facecolor='blue', arrowstyle="simple"))

plt.show()
````

````python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create plot
plt.plot(x, y, color='purple', marker='o')

# Add title and labels
plt.title("Customized Annotations")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Add customized annotation
plt.annotate("Highest Point", xy=(5, 25), xytext=(3, 20),
             fontsize=12, color='red', fontstyle='italic',
             bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='yellow'),
             arrowprops=dict(facecolor='green', arrowstyle="->", linewidth=2))

# Show plot
plt.show()
````
