```python
import pandas as pd

file_path = 'C:/Users/Razib/Downloads/spam1.csv'
message = pd.read_csv(
    file_path,
    sep=',',  # Comma-separated values
    encoding='latin1',  # Handle special characters
)

# Renaming the relevant columns
message.rename(columns={'v1': 'label', 'v2': 'message'}, inplace=True)
# Dropped unnecessary columns
message = message[['label', 'message']]  # Keeping only the required columns
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

corpus = []
for i in range(0, len(message)):
    review = re.sub('[^a-zA-Z]', ' ', message['message'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
 # Creating Bag Of Words model
 from sklearn.feature_extraction.text import CountVectorizer
 cv = CountVectorizer(max_features=2500,binary=True)
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
#Creating Bag Of Words model with ngram
from sklearn.feature_extraction.text import CountVectorizer
# For Binary BOW enable binary=True
cv = CountVectorizer(max_features=2000,binary=True,ngram_range=(3,3))
Y = cv.fit_transform(corpus).toarray()
```
